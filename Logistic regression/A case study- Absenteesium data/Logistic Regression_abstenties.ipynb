```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
preprocessed_data=pd.read_csv("df_preprocessed_new.csv")
```

```python
preprocessed_data.head()
```

### Here, as we are going to do logistic Regression, we should have the binary Target.So, we classify the "Absenteesims Time in Hours "  into binary. For this we can classify as Abseenties time more than 3 as '1' and less than that as '0'.

```python
target=np.where(preprocessed_data['Absenteeism Time in Hours']>3, 1 ,0)
```

```python
preprocessed_data['Target'] = target
```

```python
preprocessed_data
```

### so, we have the target we drop the Absenteeism Time in Hour column

```python
preprocessed_data=preprocessed_data.drop("Absenteeism Time in Hours",axis=1)
```

```python
preprocessed_data
```

## While performing logistic regression , we need to have equal proportion of Targets, or 45-55 % or 60-40% is also not bad

```python
df=preprocessed_data.copy()
```

```python
df['Target'].sum()
```

```python
df['Target'].sum() / len(df)
```

so, we have 45 % data as Target '1' and rest of it as '0'. So, it is 45-55 proportion , so we can proceed with the Logistic regression model

## As in regression is a supervised learning method, we need to split the data into INPUTS and TARGET.

```python
inputs=df.iloc[:,:-1]
```

```python
target=df.loc[:,"Target"]
```

### So, before procedding to the model we need to Standarized our INPUTS for better performance and high accuracy of the model

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_transformed=scaler.transform(inputs)
```

```python
inputs_transformed
```

## Splitting data into trainning and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs_transformed,target, train_size=0.8, random_state=666)
```

## Now, Finally its time to perform the logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0,solver='liblinear', C=10.0).fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.predict(X_train)
```

```python
clf.coef_
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_train, clf.predict(X_train))
```
